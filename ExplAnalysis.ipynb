```python
#loading packages
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib as plt 
```

```python
#loading dataset
df = pd.read_csv('ANBERD.csv')
df.head()
```

```python
#basic information about columns
df.info()
```

```python
#identifying countries 
df['Country'].unique()
```

```python
#identifying time 
df['Time'].unique()
```

```python
#finding null values
df.isnull().sum()
#Reference Period Code, Reference Period, Flag Codes, Flags can be dropped 
```

```python
#plotting R&D expenditure per country, top ten 
df.groupby('Country')['Value'].sum().sort_values(ascending=0)[:10].plot.bar()
```

```python
#plotting R&D expenditure per industry, top ten 
df.groupby('Industry')['Value'].sum().sort_values(ascending=0)[:10].plot.bar()
```

```python
#plotting R&D expenditure over time 
df_line = df[["Time", "Value"]].groupby("Time").sum("Value")
df_line.plot()
```

```python
#filtering for specific countries
dfKor = df[(df.Country == "Korea") ]
dfUS = df[(df.Country == "United States")]  
dfJap = df[(df.Country == "Japan")]  
```
